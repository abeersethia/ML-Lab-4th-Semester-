### 1.Find the mean, median, mode, variance, and standard deviation of a list. (import statistics)

```python
import statistics
data = [1, 3, 4, 5, 7, 9, 2]
x = statistics.mean(data)
y=statistics.median(data)
z=statistics.mode(data)
a=statistics.variance(data)
b=statistics.stdev(data)
print("Mean is :", x)
print("Median is :",y)
print("Mode is :",z)
print("Variance is :",a)
print("Standard Deviation is :",b)
```

### 2.Create a list of numbers and use list comprehension to generate a new list with the squares of each number

```python
data = [1, 3, 4, 5, 7, 9, 2]
newlist = []
for x in data:
    newlist.append(x*x)
print("New List:",newlist)
```

### 3. Write a function that takes a tuple of numbers and returns the product of all the elements.

```python
data=(1, 3, 4, 5, 7, 9, 2)
prod=1
for x in data:
    prod=prod*x
print("Product is:", prod)
```

### 4. Write a program that counts the frequency of each letter in each string and stores the results in a dictionary.

```python
string="naman"
letter_count = {}
for char in string:
    if char in letter_count:
        letter_count[char] += 1
    else:
        letter_count[char] = 1
print(letter_count)
```

## DATA1

```python
#a. Import the required libraries.
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

#b. Load the dataset.
df=pd.read_csv("C:\\Users\\NAMAN MAHESHWARI\\OneDrive\\Desktop\\FOURTH SEM\\ML\\LAB\\WEEK1\\Data1.csv")
df
```

```python
#c. Examine data information.
df.info()
```

```python
#d. Identify and analyze null values using df. isnull().
print(df.isnull().sum())
```

```python
#e. Conduct statistical analysis. (count, mean, SD, min, max, Quartile)
print(df.describe())
```

```python
#f. Generate boxplots for each column to detect outliers.
sns.boxplot(data=df,orient="h")
```

```python
#g. Determine quartiles and remove outliers.
for i in df.columns:
    iqr=df[i].quantile(0.75)-df[i].quantile(0.25)
    max=df[i].quantile(0.75)+(1.5*iqr)
    min=df[i].quantile(0.25)-(1.5*iqr)
    for j in df[i]:
        if(j>max or j<min):
            df[i].replace(j,df[i].median(),inplace=True)
df.describe()
```

```python
#h. Calculate correlations and visualize them with a heatmap.
correlation_matrix1 = df.corr()
sns.heatmap(correlation_matrix1, annot=True)
```

```python
#i. Normalize the various features of the given datasets. (MinMaxScaler)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df)
scaler.transform(df)
```

```python
#j. Arrange correlation values for each attribute in ascending order.
b=pd.DataFrame(df.corr())
for i in b.columns:
    print(b[i].sort_values(ascending=True))
```

```python
#j. Arrange correlation values for each attribute in ascending order.
b=pd.DataFrame(df.corr())
for i in b.columns:
    print(b[i].sort_values(ascending=True))
```
