#### Importing Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
from numpy import random
```

### Manual Method on Given Dataset

```python
x=np.array([1,2,3,4,5])
y=np.array([1.2,1.8,2.6,3.2,3.8])
plt.scatter(x,y)
```

```python
x_mean = x.mean()
y_mean = y.mean()
x_sq_mean = (x**2).mean()
xy_mean = (x*y).mean()
```

```python
a_1 = (xy_mean - x_mean*y_mean)/(x_sq_mean-x_mean**2)
a_0 = y_mean - a_1*x_mean
a = [a_0,a_1]
print(a)
```

```python
plt.scatter(x, y, color = "m",marker = "o", s = 30)
y_pred = a[0] + a[1]*x
plt.plot(x, y_pred, color = "g")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y, 'ro')
plt.axis([0, 6, 0, 5])
for xy in zip(x, y):
    plt.annotate('(%.2f, %.2f)' % xy, xy=xy)
plt.show()
```

### Inbuilt Libraries on Given Dataset

```python
df = pd.read_csv('Book1.csv')
x = df['Week'].to_numpy().reshape(-1,1)
y = df['Sales in Thousands'].to_numpy()
reg = LinearRegression().fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x)
```

```python
print(f"predicted response:\n{y_pred}")
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred, color = "g")
plt.xlabel('x')
plt.ylabel('y')
for i_x, i_y in zip(x, y):
    plt.text(i_x, i_y, '({}, {})'.format(i_x, i_y))
```
